```python
#Import pandas for data manipulation and analysis
import pandas as pd

# Load the sales dataset into a pandas DataFrame
df = pd.read_csv('sales.csv',encoding='latin1')

# Display the first 5 rows of the dataset to get a quick look
df.head()
```

```python
# Show summary of dataset: column names, types, and non-null counts
df.info()
```

```python
# Generate descriptive statistics for numeric columns
df.describe()
```

```python
# List all column names in the dataset
df.columns
```

```python
# Check how many missing values exist in each column
df.isnull().sum()
```

```python
# Calculate total sales
df['Sales'].sum()
```

```python
# Group sales by category to see which categories generate the most revenue
df.groupby('Category')['Sales'].sum()

# Total sales by Region
df.groupby('Region')['Sales'].sum()

#Category vs Region
df.groupby(['Category', 'Region'])['Sales'].sum()
```

```python
# sort based top and bottom
df.sort_values(by='Sales', ascending=False).head(5)
```

```python
# Group sales by category to the mean
df.groupby('Category')['Sales'].mean()
```

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
# Plot a bar chart showing total sales per category

# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the plots
sns.set_style('darkgrid')

# Create a bar chart
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')

# Plot histogram to see the distribution of Sales
sns.histplot(df['Sales'], bins=20)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')

# Correlation
df_numeric = df.select_dtypes(include=['number'])
correlation_matrix = df_numeric.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')


plt.show()
```
